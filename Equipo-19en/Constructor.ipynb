```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

data = pd.read_csv("ConstructorInmobiliario.csv")
data.head()
```

```python
# 1. Determinar las variables independientes
X = data[['Income', 'Size', 'Senior', 'Education']]
X 
```

```python
 #2. Determinar la variable dependiente
y = data[["SqrFt"]]
y
```

```python
# Ajustar el modelo de regresión múltiple
model = LinearRegression()
model.fit(X, y) #Entrenando el modelo
y_pred = model.predict(X) #Predicciones del modelo
y_pred

#Métricas estadísticas
mse = mean_squared_error(y, y_pred) # Error de la media cuadrática
r2 = r2_score(y, y_pred) # Varianza
```

```python
# 3. Valor de Intersección
model.intercept_
```

```python
# Coeficientes de la regresión
print("Coeficientes: ", model.coef_)
print("Error de la media cuadrática : ", mse)
print("Varianza: ", r2)
```

```python
# 4. Estimación de Tamaño de una Casa
Y = model.predict(np.array([[99,5,1,8]]))
print("Tamaño estimado de la casa: ", Y)
```

# Conclusión 
Se busco predecir el tamaño de una casa a partir de variables socioeconómicas como ingreso, educación y tamaño del hogar. El modelo permitió identificar como las variables influyen en la predicción y comprobar que la regresión lineal es una herramienta útil para estimar valores continuos a partir de datos reales.
