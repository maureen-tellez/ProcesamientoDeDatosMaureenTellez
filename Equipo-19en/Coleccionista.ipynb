```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

data = pd.read_csv("Coleccionista.csv")
data.head()
```

```python
#1. Variables independientes
X = data[['Bidders', 'Age']]
X 
```

```python
 #2. Determinar la variable dependiente
y = data[['Price']]
y
```

```python
#Ajustar el modelo de regresión múltiple
model = LinearRegression()
model.fit(X,y) #entrenar el modelo
y_pred = model.predict(X)
#y_pred

#Métricas estadísticas 
mse = mean_squared_error(y,y_pred) #error de la ,edia cuadratroca
r2= r2_score(y,y_pred) #varianza
```

```python
# #3. valor de intersección
model.intercept_
```

```python
#coeficientes 
print('Coeficientes:' , model.coef_)
print('Error cuadrático de la media: ' , mse)
print ('Varianza: ', r2)
```

```python
# 4. Estimación de precio de subasta
Y = model.predict(np.array([[14,80]]))
print("Precio estimado de una subasta: ", Y)
```
