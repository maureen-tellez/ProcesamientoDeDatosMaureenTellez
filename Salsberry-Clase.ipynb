# Regresión lineal Múltiple - SalsBerry 
1. Determinar las variables independientes
2. Determinar la variable de intersección
3. Obtener el valor de intersección
4. Estimar el corto de calefacción para una temperatura media de 30°, 5 pulgadas de aislamiento y un calentador con antigüedad de 10 años
   

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
#importar datos 
df = pd.read_csv("SalsberryRealty.csv")
df.head()
```

```python
#1. Determinar las variables independientes

X= df[['Temp','Insul','Age']]
X
```

```python
# 2.Determinar la variable dependiente
y = df[['Cost']]
y
```

```python
#Ajustar el modelo de regresión múltiple
model = LinearRegression()
model.fit(X,y) #entrenar el modelo
y_pred = model.predict(X)
#y_pred

#Métricas estadísticas 
mse = mean_squared_error(y,y_pred) #error de la ,edia cuadratroca
r2= r2_score(y,y_pred) #varianza
```

3. Valor de intersección

```python
model.intercept_
```

```python
#coeficientes 
print('Coeficientes:' , model.coef_)
print('Error cuadrático de la media: ' , mse)
print ('Varianza: ', r2)
```

4. Estimar para 30°, aislamiento 5, antigüedad de 10 años

```python
Y = model.predict([[30,5,10]])
print("El costo de la calefacción estimado es:", Y)
```
